### Importación de datos



```python
import pandas as pd # Importa la librería pandas para trabajar con DataFrames.

# Define las URLs de los archivos CSV para cada tienda.
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Carga los datos de cada URL en DataFrames separados.
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Muestra las primeras filas del primer DataFrame para una vista previa.
tienda.head()
```

#1. Análisis de facturación



# Facturación total por tienda


```python
# Calcular facturación total por tienda
facturacion_store1 = tienda["Precio"].sum() # Suma la columna 'Precio' para obtener la facturación de la tienda 1.
facturacion_store2 = tienda2["Precio"].sum() # Suma la columna 'Precio' para obtener la facturación de la tienda 2.
facturacion_store3 = tienda3["Precio"].sum() # Suma la columna 'Precio' para obtener la facturación de la tienda 3.
facturacion_store4 = tienda4["Precio"].sum() # Suma la columna 'Precio' para obtener la facturación de la tienda 4.

# Mostrar resultados de la facturación total por tienda.
print(f"Facturación TIENDA 1: ${round(facturacion_store1, 2)}")
print(f"Facturación TIENDA 2: ${round(facturacion_store2, 2)}")
print(f"Facturación TIENDA 3: ${round(facturacion_store3, 2)}")
print(f"Facturación TIENDA 4: ${round(facturacion_store4, 2)}")
```

```python
import matplotlib.pyplot as plt # Importa la librería matplotlib.pyplot para crear gráficos.


tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'] # Define una lista con los nombres de las tiendas.
facturacion = [facturacion_store1, facturacion_store2, facturacion_store3, facturacion_store4] # Define una lista con los valores de facturación calculados.

plt.figure(figsize=(8,5)) # Crea una figura y ejes para el gráfico con un tamaño específico.
plt.bar(tiendas, facturacion, color='firebrick') # Genera un gráfico de barras.
plt.title('Facturación Total por Tienda') # Establece el título del gráfico.
plt.xlabel('Tiendas') # Establece la etiqueta del eje x.
plt.ylabel('Total en pesos ($)') # Establece la etiqueta del eje y.
plt.grid(axis='y', linestyle='--', alpha=0.5) # Agrega una cuadrícula horizontal.
plt.tight_layout() # Ajusta el diseño para evitar superposiciones.
plt.show() # Muestra el gráfico.
```

# 3. Calificación promedio de la tienda


> Este paso es importante para evaluar la satisfacción del cliente, lo cual influye directamente en la fidelidad, reputación y sostenibilidad de cada tienda.



✅ Paso 1: Calcular la calificación promedio
Usamos la columna Calificación con .mean() para cada tienda

```python
# Calificación promedio por tienda
calificacion_tienda1 = tienda["Calificación"].mean()
calificacion_tienda2 = tienda2["Calificación"].mean()
calificacion_tienda3 = tienda3["Calificación"].mean()
calificacion_tienda4 = tienda4["Calificación"].mean()

# Mostrar resultados redondeados a 2 decimales
print("Calificación Promedio TIENDA 1:", round(calificacion_tienda1, 2))
print("Calificación Promedio TIENDA 2:", round(calificacion_tienda2, 2))
print("Calificación Promedio TIENDA 3:", round(calificacion_tienda3, 2))
print("Calificación Promedio TIENDA 4:", round(calificacion_tienda4, 2))
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, calificaciones, color='gold')
plt.title('Calificación Promedio por Tienda')
plt.ylabel('Promedio (escala de 1 a 5)')
plt.ylim(0, 5)
plt.xlabel('Tiendas')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

Objetivo:
Identificar qué productos son los más vendidos (alta rotación) y cuáles los menos vendidos (baja demanda) en cada tienda. Esto es clave para evaluar el rendimiento del inventario y entender la preferencia de los clientes.


##Conteo de productos vendidos por tienda
Usamos la columna Producto y el método .value_counts():




```python
# Productos más vendidos en cada tienda
productos_tienda1 = tienda["Producto"].value_counts()
productos_tienda2 = tienda2["Producto"].value_counts()
productos_tienda3 = tienda3["Producto"].value_counts()
productos_tienda4 = tienda4["Producto"].value_counts()
```

##Top 5 productos más vendidos por tienda

```python
print("-TIENDA 1 - Top 5 más vendidos:\n", productos_tienda1.head(5))
print("-TIENDA 2 - Top 5 más vendidos:\n", productos_tienda2.head(5))
print("-TIENDA 3 - Top 5 más vendidos:\n", productos_tienda3.head(5))
print("-TIENDA 4 - Top 5 más vendidos:\n", productos_tienda4.head(5))
```

##Productos menos vendidos (los que solo se vendieron una vez)



```python
# Mostrar los productos con menor número de ventas (por ejemplo, bottom 5)
print("Tienda 1 - Productos menos vendidos:\n", productos_tienda1.tail(5))
print("Tienda 2 - Productos menos vendidos:\n", productos_tienda2.tail(5))
print("Tienda 3 - Productos menos vendidos:\n", productos_tienda3.tail(5))
print("Tienda 4 - Productos menos vendidos:\n", productos_tienda4.tail(5))
```

Todas las tiendas tienen un portafolio donde cada producto fue vendido múltiples veces, lo que refleja una estrategia de ventas optimizada o una limpieza de datos previa. Sin embargo, los productos con menor frecuencia de venta podrían indicar categorías con menor rotación, lo que merece atención al tomar decisiones sobre inventario o desinversión.



# 5. Envío promedio por tienda

Objetivo:
Calcular el costo promedio de envío en cada tienda para entender su impacto en los costos operativos o en la experiencia del cliente.

Código para calcular el envío promedio
Usamos la columna Costo de envío y .mean():

```python
# Costo de envío promedio por tienda
envio_prom_tienda1 = tienda["Costo de envío"].mean()
envio_prom_tienda2 = tienda2["Costo de envío"].mean()
envio_prom_tienda3 = tienda3["Costo de envío"].mean()
envio_prom_tienda4 = tienda4["Costo de envío"].mean()

# Mostrar resultados redondeados
print("Envío Promedio TIENDA 1:", round(envio_prom_tienda1, 2))
print("Envío Promedio TIENDA 2:", round(envio_prom_tienda2, 2))
print("Envío Promedio TIENDA 3:", round(envio_prom_tienda3, 2))
print("Envío Promedio TIENDA 4:", round(envio_prom_tienda4, 2))
```

Visualización en gráfico de barras



```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
envios = [envio_prom_tienda1, envio_prom_tienda2, envio_prom_tienda3, envio_prom_tienda4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, envios, color='steelblue')
plt.title('Costo Promedio de Envío por Tienda')
plt.ylabel('Costo en pesos ($)')
plt.xlabel('Tiendas')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
